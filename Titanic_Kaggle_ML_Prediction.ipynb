```python
women = df.loc[df.Sex == 'female']["Survived"]
percent_women = sum(women)/len(women)

print(f"% of women survived is {percent_women*100}")
```

```python
men = df.loc[df.Sex == 'male']['Survived']
percent_men = sum(men)/len(men)

print(f"% of men survived is {percent_men*100}")
```

```python
cols = ['Pclass','Sex','Age','Parch','SibSp']
X_train = pd.get_dummies(df[cols])
```

```python
Y_train = df['Survived']
```

```python
test_df = pd.read_csv("https://raw.githubusercontent.com/ritzx21/Titanic-ML-prediction-Kaggle-/main/Titanic%20Dataset%20(Kaggle%20competition)/test.csv")
```

```python
test_df['Cabin'] = test_df['Cabin'].fillna(" ")
```

```python
test_df.Fare = test_df.Fare.fillna(test_df.Fare.mean())
```

```python
test_cols = ['Pclass','Sex','Age','Parch','SibSp']
X_test = pd.get_dummies(test_df[test_cols])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train , Y_train)
```

```python
Y_predict = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': Y_predict})
output.to_csv('Submission.csv', index=False)
print("Saved")
```
